```python
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
from waxwing import waxwing as wx
import numpy as np
%config InlineBackend.figure_format = 'retina'
```

default styling

```python
# scatter plot
wx.set_default_styles()

wx.set_palette(wx.palettes.matcha)
for i in range(5):
    data = np.random.normal(loc=i, scale=0.5, size=(100, 2))
    plt.scatter(data[:, 0], data[:, 1], label=f"Category {i}", s=10)
wx.style_axes()
plt.xlabel("X data (units)")
plt.show()
```

changing th figure size, font, and color palette

```python
# bar plot
wx.set_figsize(3.5, 1.5)
wx.set_palette(wx.palettes.fruit)
wx.set_font("Literata", "light")
df = sns.load_dataset("penguins")
sns.barplot(data=df, x="sex", y="body_mass_g", hue="species")
plt.ylim(0, 6000)
wx.style_axes(encompass_data=False)
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

Setquential color map, and changing the font + line weight

```python
wx.set_palette(wx.palettes.canopy)
# changing font weight also changes the line weights to match
wx.set_font("Noto Sans", "regular")
x = np.linspace(0, 10, 100)
for i in range(5):
    y = np.sin(x + i) + i
    plt.plot(x, y, label=f"Line {i}")
wx.style_axes(encompass_data=True)
plt.xlabel("X data (units)")
plt.ylabel("Y data (units)")
plt.legend().set_visible(False)
plt.show()
```
